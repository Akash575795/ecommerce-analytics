##### Import Libraries

```python
import  pandas as pd
import numpy as np
```

##### Load Data

```python
df = pd.read_csv("C:/Users/akash/Downloads/e-commerce/data/processed/ecommerce_analytics.csv")
df.shape
```

##### Data types correction

```python
df['date'] = pd.to_datetime(df['date'], dayfirst= True)
df['contact_no'] = df['contact_no'].astype(str)
df['nid'] = df['nid'].astype(str)
```

```python
df.info()
```

##### Total Revenue

```python
Total_Revenue = df['total_price'].sum()

print("Total Revenue: " , Total_Revenue)
```

##### Total Transactions

```python
Total_transactions = len(df)
print("Total Transactions:",Total_transactions)
```

##### Total Customers

```python
total_customers = df['coustomer_key'].nunique()
print("Total Customers:", total_customers)
```

##### Revenue by Year

```python
revenue_by_year = df.groupby('year')['total_price'].sum().sort_values(ascending = False)

print(revenue_by_year)
```

##### Revenue by Month

```python
revenue_by_year_month = df.groupby(['year', 'month'])['total_price'].sum().sort_index()
print( revenue_by_year_month)
```

##### Revenue by Product

```python
revenue_by_product = df.groupby('item_name')['total_price'].sum().sort_values(ascending=False)
print("Revenue by Product:", revenue_by_product)
```

##### Quantity vs Revenue

```python
df.groupby('item_name').agg(
    total_quantity=('quantity', 'sum'),
    total_revenue=('total_price', 'sum')
).sort_values('total_revenue', ascending=False).head(10)
```

##### Revenue by Customer

```python
df.groupby('coustomer_key')['total_price'].sum().sort_values(ascending=False).head(10)
```

##### Customer Purchase Frequency

```python
df.groupby('coustomer_key').size().sort_values(ascending=False).head(10)
```

##### Revenue by Division

```python
df.groupby('division')['total_price'].sum().sort_values(ascending=False)
```

##### Revenue by District

```python
df.groupby('district')['total_price'].sum().sort_values(ascending=False).head(10)
```

##### Revenue by Store

```python
df.groupby('store_key')['total_price'].sum().sort_values(ascending=False).head(10)
```
